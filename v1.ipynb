```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(x):
    return np.round(np.sin(np.radians(x)), 10)
    
```

```python
x = np.linspace(0, 4*90, 400)
f1 = 1 + f(2*x - 90)
def plot_sin_degrees():
    # Define the range of x values in degrees
    y = f1         # Calculate sin(x) in degrees

    # Create the plot
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='sin(x)', color='blue')
    plt.title('Visualization of sin(x) where x is in degrees')
    plt.xlabel('Units (each unit represents 90 degrees)')
    plt.ylabel('sin(x)')
    
    # Customize x-ticks
    ticks = np.arange(0, 5, 1)  # From -4 to 4 for 90-degree increments
    tick_labels = [str(tick) for tick in ticks]  # Create labels as strings
    plt.xticks(ticks * 90, tick_labels)  # Set ticks and labels

    plt.axhline(0, color='black', lw=0.5, ls='--')  # Add a horizontal line at y=0
    plt.axvline(0, color='black', lw=0.5, ls='--')  # Add a vertical line at x=0
    plt.grid()
    plt.legend()
    plt.ylim(0, 2.5)  # Set y-axis limits
    plt.show()

# Call the function to plot
plot_sin_degrees()
```

```python
data = [0, 0, 2, 2, 0, 0]
def g(x):
    array = data
    return array[x]
```

```python
def plot_g():
    # Define the range of x values (0 to 5) and scale to 90-degree increments
    x = np.arange(0, 6) * 90  # Scale x to match the 90-degree unit increments
    y = [g(i // 90) for i in x]  # Calculate g(x) for each x value

    # Create the plot
    plt.figure(figsize=(10, 5))

    # Plot the lines connecting the points
    plt.plot(x, y, marker='*', linestyle='--', color='red', label='g(x)', markersize=8)
    
    plt.title('Visualization of g(x)')
    plt.xlabel('Units (each unit represents 90 degrees)')
    plt.ylabel('g(x)')
    
    plt.xticks(x)  # Set x-ticks to match x values (scaled by 90 degrees)
    plt.axhline(0, color='black', lw=0.5, ls='--')  # Add a horizontal line at y=0
    plt.grid()

    plt.ylim(-1, 3)  # Set y-axis limits
    plt.legend()  # Add a legend
    plt.show()

# Call the function to plot g(x)
plot_g()
```

```python
def plot_sin_degrees():
    # Define the range of x values in degrees
    x = np.linspace(0, len(data) * 90, 400)  # 400 points from 0 to 360 degrees

    # Calculate f1 and f2
    f1 = 1 + f(2 * x - 90)  # f1 function
    f2 = 1 + f(2 * x + 90)  # f2 function

    # Create the plot
    plt.figure(figsize=(10, 5))
    
    # Plot f1
    plt.plot(x, f1, label='f1 = 1 + sin(2x - 90°)', color='blue')
    
    # Plot f2
    plt.plot(x, f2, label='f2 = 1 + sin(2x + 90°)', color='orange')

    plt.title('Visualization of f1 and f2 where x is in degrees')
    plt.xlabel('Units (each unit represents 90 degrees)')
    plt.ylabel('Function values')
    
    # Customize x-ticks
    ticks = np.arange(0, len(data) + 1, 1)  
    tick_labels = [str(tick) for tick in ticks]  # Create labels as strings
    plt.xticks(ticks * 90, tick_labels)  # Set ticks and labels

    plt.axhline(0, color='black', lw=0.5, ls='--')  # Add a horizontal line at y=0
    plt.axvline(0, color='black', lw=0.5, ls='--')  # Add a vertical line at x=0
    plt.grid()
    plt.legend()
    plt.ylim(0, 2.5)  # Set y-axis limits
    plt.show()

# Call the function to plot both f1 and f2
plot_sin_degrees()
```

```python
# Function to visualize g(x), f1(x), and f2(x)
def plot_functions():
    # Define the range of x values for g(x) and scale to match 90-degree units
    x_g = np.arange(0, 6) * 90  # Scale x_g by 90 to match the scale of f1 and f2
    y_g = [g(i // 90) for i in x_g]  # Calculate g(x) for each x

    # Define the range of x values for f1(x) and f2(x)
    x_f = np.linspace(0, len(data) * 90, 400)  # x values from 0 to 360 degrees

    # Calculate f1(x) and f2(x)
    f1 = 1 + f(2 * x_f - 90)  # f1(x) = 1 + sin(2x - 90)
    f2 = 1 + f(2 * x_f + 90)  # f2(x) = 1 + sin(2x + 90)

    # Create the plot
    plt.figure(figsize=(12, 6))

    # Plot g(x) (scaled by 90 degrees)
    plt.plot(x_g, y_g, marker='o', linestyle='-', color='orange', label='g(x)', markersize=8)

    # Plot f1(x)
    plt.plot(x_f, f1, color='blue', label='f1(x) = 1 + sin(2x - 90)', linewidth=2)

    # Plot f2(x)
    plt.plot(x_f, f2, color='green', label='f2(x) = 1 + sin(2x + 90)', linewidth=2)

    # Customize the plot
    plt.title('Visualization of g(x), f1(x), and f2(x)')
    plt.xlabel('Units (each unit represents 90 degrees)')
    plt.ylabel('Function Values')
    plt.axhline(0, color='black', lw=0.5, ls='--')  # Horizontal line at y=0
    plt.grid()
    plt.legend()  # Add a legend
    plt.ylim(-1, 3)  # Set y-axis limits
    plt.show()

# Call the function to plot g(x), f1(x), and f2(x)
plot_functions()
```

```python
x_g = np.arange(0, len(data)) * 90  
y_g = [g(i // 90) for i in x_g]


def checker(f1, f2):
    alpha = int(max(f1, f2) - min(f1, f2))
    
    shouldCount = False
    counter = 0
    
    for i in range(len(x_g)):
        if x_g[i] % 360 == 0:
            x_g[i] = 0
            
        current_f1 = 1 + f(2 * x_g[i] - 90)
        current_f2 = 1 + f(2 * x_g[i] + 90)
        
        if current_f1 == y_g[i] or current_f2 == y_g[i]:
            if i < len(x_g) - 1 and y_g[i] == min(current_f1, current_f2) and y_g[i + 1] == max(current_f1, current_f2):
                counter += 1
                shouldCount = True
    
            if i < len(x_g) - 1 and y_g[i] == max(current_f1, current_f2) and y_g[i + 1] == min(current_f1, current_f2):
                counter += 1
                shouldCount = False
    
            if shouldCount:
                counter += 1
    
    return [alpha, counter]
```

```python
# Renamed variable to avoid shadowing the built-in function max
max_count_value = 0  
alpha_value = 0  

# Make sure you have a valid loop for 'i' in this context
maxOfY = max(y_g)
minOfY = min(y_g)

 
for i in range(len(x_g)):
    for p in range(minOfY, maxOfY, 1):
        for m in range(1, maxOfY, 1):
            if abs(p) + m > 10:
                break
            f1 = p + m * f(2 * x_g[i] - 90)
            f2 = p + m * f(2 * x_g[i] + 90)

            checker_result = checker(f1, f2)  # Store result of checker
            if checker_result[1] > max_count_value:  # Compare against the new variable
                max_count_value = checker_result[1]
                alpha_value = checker_result[0]

print("The alpha is ", alpha_value," and it appears ",max_count_value," times!!")
```
